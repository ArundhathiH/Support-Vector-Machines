# 1. Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE, SVMSMOTE
import warnings
warnings.filterwarnings('ignore')
```

# 2. Load Dataset

```python
pd.set_option('display.max_columns',None)
ff_data=pd.read_csv('forestfires.csv')
ff_data
```

# 3. EDA

```python
ff_data.info()
```

```python
ff_data.describe()
```

```python
ff_data['size_category'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(ff_data['size_category'].value_counts(),labels=['Small area','Large area'],explode=[0.02,0.02],autopct='%1.2f%%',textprops={'fontsize':14},colors=['plum','salmon'])
plt.show()
```

# 4. Data Preprocessing and Data Preparation

```python
ff_data.drop(['month','day'],axis=1,inplace=True)
```

```python
le=LabelEncoder()
ff_data['size_category']=le.fit_transform(ff_data['size_category'])
```

```python
ff_data
```

# 5. Model Building

```python
X=ff_data.drop('size_category',axis=1)
y=ff_data[['size_category']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

# 6. Model Training

```python
svm=SVC()
svm.fit(X_train,y_train)
```

# 7. Model Testing

```python
y_pred=svm.predict(X_test)
```

# 8. Model Evaluation

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
svm.score(X_train,y_train)
```

# 9. Imbalance Data Handling

```python
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
X_res
```

```python
y_res
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.2,random_state=10)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

# 10. Tweaking of Hyperparameters

```python
svm=SVC(C=2,gamma=0.001)
svm.fit(X_train,y_train)
```

# Model Testing

```python
y_pred=svm.predict(X_test)
```

# Model Evaluation

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

### Train Accuracy

```python
round(svm.score(X_train,y_train),3)
```

### Test Accuracy

```python
round(svm.score(X_test,y_test),3)
```
